```python
import numpy as np
```

```python
from matplotlib import pyplot as plt
```

```python
x=np.arange(1,11)
```

```python
x
```

```python
print(x)
```

```python
y=2*x
```

```python
y
```

<h1 style="color: PINK; font-size: 40px; font-weight: bold;">LINE CHART</h1>

```python
plt.plot(x,y)
plt.title("Line Chart")    
plt.xlabel("X Axis")       
plt.ylabel("Y Axis")     

plt.show()
```

<h1 style="color: SKYBLUE; font-size: 40px; font-weight: bold;">BAR CHART</h1>

```python
plt.bar(x,y)
plt.title("Line Chart")    
plt.xlabel("X Axis")       
plt.ylabel("Y Axis")     

plt.show()
```

```python
x = np.arange(1, 11)                 # x values (1 to 10)
y = np.random.randint(1, 100, 10)    # random integers for y

# Create bar chart
plt.bar(x, y, color='orange', edgecolor='black')

# Add labels and title
plt.title("Random Bar Chart")
plt.xlabel("X Axis")
plt.ylabel("Y Axis (Random Values)")

# Show the chart
plt.show()
```

<h1 style="color: orange; font-size: 40px;">SCATTER PLOT</h1>

```python
plt.scatter(x,y,color='red',edgecolor='green',s=200)
plt.title("Scatter Chart")    
plt.xlabel("X Axis")       
plt.ylabel("Y Axis")     

plt.show()
```

<h1 style="color: red; font-size: 40px; font-weight: bold;">HISTOGRAM</h1>


```python
data = np.random.randn(1000)

# Create histogram
plt.hist(data, bins=30, color='pink', edgecolor='black')

# Add labels and title
plt.title("Histogram Example")
plt.xlabel("Value")
plt.ylabel("Frequency")

# Show plot
plt.show()
```

<h1 style="color: red; font-size: 40px; font-weight: bold;">BOX PLOT</h1>

```python
plt.boxplot(x)
plt.show()
```
